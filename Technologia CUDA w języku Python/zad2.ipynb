```python
import torch
import time
torch.cuda.is_available()
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
A = torch.randint(0, 2, (2, 2), device=device)
B = torch.randint(2, 4, (2, 2), device=device)

prog = 2
row_sums = A.sum(dim=1)
mask = row_sums >= prog

result = torch.mm(A[mask].float(), B.float()).to(torch.long)

print(A)
print(B)
print(result)
```

```python
size=10000
A_cpu = torch.randn(size, size)
start_cpu = time.time()
res_cpu = torch.mm(A_cpu.t(), A_cpu)
end_cpu = time.time()
print(f"CPU {end_cpu - start_cpu}")

A_gpu = torch.randn(size, size, device=device)

start_gpu = time.time()
res_gpu = torch.mm(A_gpu.t(), A_gpu)
torch.cuda.synchronize()
end_gpu = time.time()

print(f"GPU {end_gpu - start_gpu}")
```
